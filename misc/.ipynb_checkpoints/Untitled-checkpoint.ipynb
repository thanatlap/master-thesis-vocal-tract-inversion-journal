```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
import librosa.display
import librosa
```

```python
sound = librosa.load('../data/d_dataset_2/sound/sound0.wav', sr=16000)[0]
```

```python
t=np.linspace(0, len(sound)/16000, num=len(sound))
plt.figure(1)
plt.title("Signal Wave...")
plt.plot(t, sound)
plt.show()
```

```python
sound2 = librosa.load('../data/d_dataset_2/sound/sound11.wav', sr=16000)[0]
t=np.linspace(0, len(sound2)/16000, num=len(sound2))
plt.figure(1)
plt.title("Signal Wave...")
plt.plot(t, sound2)
plt.show()
```

```python
sound2_pad = np.pad(sound2, (0, max(0, len(sound) - sound2.shape[0])), "constant")
t=np.linspace(0, len(sound)/16000, num=len(sound))
plt.figure(1)
plt.title("Signal Wave...")
plt.plot(t, sound2_pad)
plt.show()
```

```python
mfcc = librosa.feature.mfcc(sound2_pad, sr=16000, n_mfcc = 13)[:13]
```

```python
librosa.display.specshow(mfcc, x_axis='time')
```

```python
mfcc1 = librosa.feature.mfcc(sound, sr=16000, n_mfcc = 13)
librosa.display.specshow(mfcc1, x_axis='time')
```
