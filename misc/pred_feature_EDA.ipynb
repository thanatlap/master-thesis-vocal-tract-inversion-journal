```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
import librosa.display
import librosa
from librosa import feature
from IPython import display
from matplotlib import cm
from functools import partial
np_load_old = partial(np.load)
np.load = lambda *a,**k: np_load_old(*a, allow_pickle=True, **k)

%matplotlib inline
```

```python
mfcc_record = np.load('../data/d_records/d_record_sorn_sis/prep_data_13/features.npy')
mfcc_record_1 = np.load('../data/d_records/selected/prep_data_13/features.npy')
```

```python
def plot_mfcc(mfcc, idx=1):
    fig, ax = plt.subplots(figsize=(10,5))
    cax = ax.imshow(np.swapaxes(mfcc[idx], 0 ,1)[:13], interpolation='nearest', cmap=cm.coolwarm, origin='lower')
    ax.set_title('MFCC')

    plt.show()
```

```python
plot_mfcc(mfcc_record, idx=2)
```

```python
plot_mfcc(mfcc_record_1, idx=2)
```
